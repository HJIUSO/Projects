```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
#read the file
math = pd.read_csv("student_math_clean.csv")
```

```python
math
```

```python
#getting the shape of the data set
math.shape
```

```python
#getting the size of the data set
math.size
```

```python
#checking for null value in the data set
math.isnull()
```

```python
#checking for null value in the data set
math.isnull().any()
```

```python
#creating a new data frame only contain int64 type
num = math.select_dtypes("int64")
num.shape
```

```python
#function to create a new data frame with student who have grade = 15 (above 75%)
top = math[math["final_grade"] >= 15]
top.shape
```

```python
#ploting a heatmap to show the correlation between each integer value in the data set
plt.figure(figsize=(16,8))
sns.heatmap(num.corr(), annot=True, cmap ='twilight')
```

```python
#create boxplots for the final grades based on gender, class failure, and age
plt.figure(figsize=(15,5))

plt.subplot(1,3,1)
sns.boxplot(y=math['final_grade'], x= math['sex'])
plt.title('Final Grade by Sex overall')

plt.subplot(1,3,2)
sns.boxplot(y=math['final_grade'], x= math['class_failures'])
plt.title('Final Grade by Class Failures overall')

plt.subplot(1,3,3)
sns.boxplot(y= math['final_grade'], x=math['age'])
plt.title('Final Grade by Age overall')
```

```python
#create boxplots for the final grades based on parent status, address type, and family size
plt.figure(figsize=(15,5))

plt.subplot(1,3,1)
sns.boxplot(y=math['final_grade'], x= math['parent_status'])
plt.title('Final Grade by Parent Status overall')

plt.subplot(1,3,2)
sns.boxplot(y=math['final_grade'], x= math['address_type'])
plt.title('Final Grade by Address Type overall')

plt.subplot(1,3,3)
sns.boxplot(y= math['final_grade'], x=math['family_size'])
plt.title('Final Grade by Family Size overall')
```

```python
#create boxplots for the final grades 75% and above based on gender, class failure, and age
plt.figure(figsize=(15,5))

plt.subplot(1,3,1)
sns.countplot( x= top['final_grade'], hue = top['sex'])
plt.legend(loc='best')
plt.title('Final Grade of 75% and above by Sex')

plt.subplot(1,3,2)
sns.countplot(x= top['final_grade'], hue= top['class_failures'])
plt.legend(loc='best')
plt.title('Final Grade of 75% and above by Class Failures')

plt.subplot(1,3,3)
sns.countplot(x= top['final_grade'], hue=top['age'])
plt.legend(loc='best')
plt.title('Final Grade of 75% and above by Age')
```

```python
#create boxplots for the final grades 75% and above based on parent status, address type, and family size
plt.figure(figsize=(15,5))

plt.subplot(1,3,1)
sns.countplot( x= top['final_grade'], hue = top['parent_status'])
plt.legend(loc='best')
plt.title('Final Grade of 75% and above by Parent Status')

plt.subplot(1,3,2)
sns.countplot(x= top['final_grade'], hue= top['address_type'])
plt.legend(loc='best')
plt.title('Final Grade of 75% and above by Address Type')

plt.subplot(1,3,3)
sns.countplot(x= top['final_grade'], hue=top['family_size'])
plt.legend(loc='best')
plt.title('Final Grade of 75% and above by Family Size')
```

```python
#function to create a new data frame with student who have grade < 10(below average)
low = math[math["final_grade"] < 10]
low.shape
```

```python
#create boxplots for the final grades below average based on gender and age
plt.figure(figsize=(15,5))

plt.subplot(1,2,1)
sns.countplot( x= low['final_grade'], hue = low['sex'])
plt.legend(loc='best')
plt.title('Final Grade of Below Average by Sex')

plt.subplot(1,2,2)
sns.countplot(x= low['final_grade'], hue=low['age'])
plt.legend(loc='best')
plt.title('Final Grade of Below Average by Age')
```

```python
#create boxplots for the final grades below average based on parent status, address type, and family size
plt.figure(figsize=(15,5))

plt.subplot(1,3,1)
sns.countplot( x= low['final_grade'], hue = low['parent_status'])
plt.legend(loc='best')
plt.title('Final Grade of Below Average by Parent Status')

plt.subplot(1,3,2)
sns.countplot(x= low['final_grade'], hue= low['address_type'])
plt.legend(loc='best')
plt.title('Final Grade of Below Average by Address Type')

plt.subplot(1,3,3)
sns.countplot(x= low['final_grade'], hue=low['family_size'])
plt.legend(loc='best')
plt.title('Final Grade of Below Average by Family Size')
```

```python
#plot for final grades based on Study time
#this plot is created to strengthen a hypothesis (not use for the report)
plt.figure(figsize=(15,5))

plt.subplot(1,3,1)
sns.boxplot(y=math['final_grade'], x= math['study_time'])
plt.title('Final Grade by Study Time overall')

plt.subplot(1,3,2)
sns.countplot(x= top['final_grade'], hue = math['study_time'])
plt.title('Final Grade by Study Time 75% and Above')

plt.subplot(1,3,3)
sns.countplot(x=low['final_grade'],  hue= math['study_time'])
plt.title('Final Grade by Study Time Below Average')
```
